# cylinders and model year

```python
a=df.groupby(['model_year'],as_index=False)['cylinders'].sum()
b=sns.barplot(x='model_year',y='cylinders',data=a)
for bars in b.containers:
    b.bar_label(bars)
```

```python
#from this we can tell that in the year 73 most of the cylinders are been sold and in the year 80 less cylinders are been sold.
```

# weight vs model year

```python
a=df.groupby(['model_year'],as_index=False)['weight'].sum()
b=sns.barplot(x='model_year',y='weight',data=a)
for bars in b.containers:
    b.bar_label(bars)
```

```python
#from this we can tell that since many cylinders are been sold in the year 73 so the weight of the cylinders are also high in the 
#year , similarly in the year 80 less productio  of cylinders took place so the weight of cylinders are less in year 80.
```

# model_year vs mpg

```python
a=df.groupby(['model_year'],as_index=False)['mpg'].sum()
b=sns.barplot(x='model_year',y='mpg',data=a)
for bars in b.containers:
    b.bar_label(bars)
```

```python
#from this we can tell that since in the year 80 less production of cylinders took place so due to this the mpg was more in the 
#year of 80 from this we can conclude that there is an inverse relationship between mpg and weight,cylinders
```

# model_year vs acceleration

```python
a=df.groupby(['model_year'],as_index=False)['acceleration'].sum()
b=sns.barplot(x='model_year',y='acceleration',data=a)
for bars in b.containers:
    b.bar_label(bars)
```

```python
#from this we can tell that since weight and cyclinders are increasing so dut to this the acceleration also increases so 
#from this graph we can tell that in the year 78 acceleration was max because of the rise of cylinders and weight .
#there is a inverse relationship between acceleration , cylinders and weight and we have direct relationship with mpg
```

# model year vs displacement

```python
a=df.groupby(['model_year'],as_index=False)['displacement'].sum()
b=sns.barplot(x='model_year',y='displacement',data=a)
for bars in b.containers:
    b.bar_label(bars)
```

```python
#from this we can tell that since weight and cyclinders are increasing so dut to this the acceleration also increases so
#due to this reason displacement also increases in the year 73 as in the year 78 acceleration was max because of the fall of cylinders and weight .
#So there is a inverse relationship between displacement ,weight ,cylinders and direct relationship with mpg and acceleration
```

# model_year vs horsepower

```python
a=df.groupby(['model_year'],as_index=False)['horsepower'].sum()
b=sns.barplot(x='model_year',y='horsepower',data=a)
for bars in b.containers:
    b.bar_label(bars)
```
